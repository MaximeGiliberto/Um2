```python
# -*- coding: utf-8 -*-


import sqlite3

# Connect to the database
conn = sqlite3.connect('D:\\UTC\\Info\\US01\\Hackathon\\UTSEUS-shanghai-dianping.db')
c1 = conn.cursor()
c2 = conn.cursor()


# SQL request         
c1.execute('SELECT name, address, latitude, longitude, avg_rating, avg_price, category '+
          'FROM venues, venue_categories '+
          'WHERE venues.business_id = venue_categories.business_id '+
          'AND venue_categories.category = "电影院" ')

venues1 = c1.fetchall()

c2.execute('SELECT name, address, latitude, longitude, avg_rating, avg_price, category '+
          'FROM venues, venue_categories '+
          'WHERE venues.business_id = venue_categories.business_id '+
          'AND venue_categories.category = "家装卖场" ')

venues2 = c2.fetchall()



```

```python
# Convert into a pandas dataframe
import pandas as pd

labels = ['name', 'address', 'latitude', 'longitude', 'avg_rating', 
          'avg_price', 'category']

df1 = pd.DataFrame.from_records(venues1, columns=labels)
df2 = pd.DataFrame.from_records(venues2, columns=labels)

df1.to_csv('Cinema.csv')
df2.to_csv('FurnitureShop.csv')



print(df1)
print(df2)
```
